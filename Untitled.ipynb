```python

import os
import cv2
import numpy as np
from tqdm import tqdm
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from tensorflow.keras.utils import to_categorical
import tensorflow as tf
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, TensorDataset
import torchvision.models as models
import torchvision.transforms as transforms
print("Setup complete!")
```

```python
# Setting paths for Colab
IMGSIZE = 150
BASEDIR = '/content/drive/MyDrive/DL_Project_3/flowers'
FLOWER_DAISY_DIR = os.path.join(BASEDIR, 'daisy')
FLOWER_DANDELION_DIR = os.path.join(BASEDIR, 'dandelion')
FLOWER_ROSE_DIR = os.path.join(BASEDIR, 'rose')
FLOWER_SUNFLOWER_DIR = os.path.join(BASEDIR, 'sunflower')
FLOWER_TULIP_DIR = os.path.join(BASEDIR, 'tulip')

print("Dataset ready!")

print("Daisy dir:     ", FLOWER_DAISY_DIR)
print("Dandelion dir: ", FLOWER_DANDELION_DIR)
print("Rose dir:      ", FLOWER_ROSE_DIR)
print("Sunflower dir: ", FLOWER_SUNFLOWER_DIR)
print("Tulip dir:     ", FLOWER_TULIP_DIR)
```

```python
# Containers for images and labels
X = []
Z = []

def assign_label(img, flowertype):
    return flowertype

def make_train_data(flowertype, DIR):
    for img in tqdm(os.listdir(DIR), desc=f"Loading {flowertype}"):
        path = os.path.join(DIR, img)
        if not os.path.isfile(path): continue
        img_arr = cv2.imread(path, cv2.IMREAD_COLOR)
        if img_arr is None: continue
        img_arr = cv2.resize(img_arr, (IMGSIZE, IMGSIZE))
        X.append(np.array(img_arr))
        Z.append(str(assign_label(img, flowertype)))

# Load all flowers
make_train_data('daisy', FLOWER_DAISY_DIR)
make_train_data('dandelion', FLOWER_DANDELION_DIR)
make_train_data('rose', FLOWER_ROSE_DIR)
make_train_data('sunflower', FLOWER_SUNFLOWER_DIR)
make_train_data('tulip', FLOWER_TULIP_DIR)

print(f"Total images loaded: {len(X)}")
print(f"Total labels loaded: {len(Z)}")
```

```python
# Processing the raw image lists into standardized Numpy arrays
# Converted list of images to NumPy array and normalizing to [0,1]
# Converted to numpy array and normalize
X = np.array(X, dtype=np.float32) / 255.0
print(f"X shape: {X.shape}")

# Label encoding
le = LabelEncoder()
Y_int = le.fit_transform(Z)
print("Unique classes and indices:", list(zip(le.classes_, range(len(le.classes_)))))

# One-hot encode
Y = to_categorical(Y_int, num_classes=5)
print(f"Y_int shape: {Y_int.shape}")
print(f"Y one-hot shape: {Y.shape}")
```

```python
# Train/test split according to the requirement 75/25
x_train, x_test, y_train, y_test = train_test_split(
    X, Y, test_size=0.25, random_state=42, shuffle=True, stratify=Y_int
)
print(f"x_train.shape: {x_train.shape}")
print(f"y_train.shape: {y_train.shape}")
print(f"x_test.shape: {x_test.shape}")
print(f"y_test.shape: {y_test.shape}")

```

```python
import tensorflow as tf

# Resizing training and test images to 224x224 for ResNet50

IMG_RESIZE = 224
x_train_resized = tf.image.resize(x_train, [IMG_RESIZE, IMG_RESIZE])
x_test_resized = tf.image.resize(x_test, [IMG_RESIZE, IMG_RESIZE])
print(f"x_train_resized shape: {x_train_resized.shape}")
print(f"x_test_resized shape: {x_test_resized.shape}")

# Load pretrained ResNet50
base_model = tf.keras.applications.ResNet50(
    weights='imagenet',
    include_top=False,
    input_shape=(IMG_RESIZE, IMG_RESIZE, 3)
)

# Freeze all base model layers
for layer in base_model.layers:
    layer.trainable = False

base_model.summary()

```

```python
# Adding custom classification head
x = base_model.output
x = tf.keras.layers.GlobalAveragePooling2D()(x)  # Reduces (7×7×2048) → (2048,) feature vector
x = tf.keras.layers.Dense(512, activation='relu')(x)  # Hidden layer for flower-specific learning
x = tf.keras.layers.Dropout(0.5)(x) # Prevents overfitting
outputs = tf.keras.layers.Dense(5, activation='softmax')(x)

model_tf = tf.keras.Model(inputs=base_model.input, outputs=outputs)
model_tf.summary()


```

```python
# Compile with lower learning rate for transfer learning
# Adam optimizer and Low learning rate prevents destroying pretrained ImageNet weights
# The learning rate is 0.0001
model_tf.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=1e-4),
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

```

```python
BATCH_SIZE = 32  # Batch Size 32 for memory efficiency
EPOCHS = 20

history_tf = model_tf.fit(
    x_train_resized, y_train,
    validation_data=(x_test_resized, y_test),
    epochs=EPOCHS,
    batch_size=BATCH_SIZE,
    verbose=1
)
# Uses resized train data (3237 images), validates on test set (1080 images)
```

```python
# Evaluate
# After model_tf.fit(...)
val_loss_tf = history_tf.history['val_loss']
val_acc_tf  = history_tf.history['val_accuracy']
train_loss_tf = history_tf.history['loss']
train_acc_tf  = history_tf.history['accuracy']

print("TF final  loss:", train_loss_tf[-1])
print("TF final  acc :", train_acc_tf[-1])
print("TF final val loss  :", val_loss_tf[-1])
print("TF final val acc   :", val_acc_tf[-1])
```

```python
import numpy as np
from sklearn.metrics import classification_report

# y_test is one‑hot; x_test_resized was used for evaluation
y_pred_probs = model_tf.predict(x_test_resized)
y_pred_classes = np.argmax(y_pred_probs, axis=1)
y_true_classes = np.argmax(y_test, axis=1)
```

```python
target_names = ['daisy', 'dandelion', 'rose', 'sunflower', 'tulip']

print("TensorFlow model – classification report:")
print(classification_report(y_true_classes,
                            y_pred_classes,
                            target_names=target_names))
```

```python
# Tensorflow data to pytorch format and creating data loaders
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import TensorDataset, DataLoader

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Using device:", device)

# Convert from NHWC (TensorFlow) to NCHW (PyTorch)
x_train_torch = torch.from_numpy(x_train_resized.numpy()).permute(0, 3, 1, 2)  # (N, 3, 224, 224)
x_test_torch  = torch.from_numpy(x_test_resized.numpy()).permute(0, 3, 1, 2)

# For labels, convert from one-hot to class indices 0..4
y_train_indices = torch.from_numpy(np.argmax(y_train, axis=1)).long()
y_test_indices  = torch.from_numpy(np.argmax(y_test, axis=1)).long()

print("x_train_torch shape:", x_train_torch.shape)
print("y_train_indices shape:", y_train_indices.shape)

# Create TensorDataset and DataLoader
train_dataset = TensorDataset(x_train_torch, y_train_indices)
test_dataset  = TensorDataset(x_test_torch,  y_test_indices)

train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
test_loader  = DataLoader(test_dataset,  batch_size=32, shuffle=False)
```

```python
import torchvision.models as models

# Load pre-trained ResNet50
resnet = models.resnet50(weights=models.ResNet50_Weights.IMAGENET1K_V1)

# Number of input features to the original fc layer
num_features = resnet.fc.in_features

# Replacing the fully connected layer with a new classifier for 5 classes
resnet.fc = nn.Sequential(
    nn.Linear(num_features, 512),
    nn.ReLU(),
    nn.Dropout(0.5),
    nn.Linear(512, 5)
)

# Freeze all layers except the classifier
for name, param in resnet.named_parameters():
    if not name.startswith("fc."):
        param.requires_grad = False

resnet = resnet.to(device)

print(resnet)  # Saving for report
```

```python
# The learning rate is 0.0001
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(filter(lambda p: p.requires_grad, resnet.parameters()), lr=1e-4)
```

```python
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(filter(lambda p: p.requires_grad, resnet.parameters()), lr=1e-4)

EPOCHS = 20
history_pt = {'train_loss': [], 'train_acc': [], 'val_loss': [], 'val_acc': []}

for epoch in range(EPOCHS):
    # ---- Train ----
    resnet.train()
    running_loss = 0.0
    correct = 0
    total = 0

    for inputs, labels in train_loader:
        inputs = inputs.to(device)
        labels = labels.to(device)

        optimizer.zero_grad()
        outputs = resnet(inputs)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        running_loss += loss.item() * inputs.size(0)
        _, predicted = torch.max(outputs, 1)
        total += labels.size(0)
        correct += (predicted == labels).sum().item()

    train_loss = running_loss / total
    train_acc = correct / total

    # ---- Validation ----
    resnet.eval()
    val_running_loss = 0.0
    val_correct = 0
    val_total = 0
    with torch.no_grad():
        for inputs, labels in test_loader:
            inputs = inputs.to(device)
            labels = labels.to(device)

            outputs = resnet(inputs)
            loss = criterion(outputs, labels)

            val_running_loss += loss.item() * inputs.size(0)
            _, predicted = torch.max(outputs, 1)
            val_total += labels.size(0)
            val_correct += (predicted == labels).sum().item()

    val_loss = val_running_loss / val_total
    val_acc = val_correct / val_total

    history_pt['train_loss'].append(train_loss)
    history_pt['train_acc'].append(train_acc)
    history_pt['val_loss'].append(val_loss)
    history_pt['val_acc'].append(val_acc)

    print(f"Epoch {epoch+1}/{EPOCHS} "
          f"- Train Loss: {train_loss:.4f} - Train Acc: {train_acc:.4f} "
          f"- Val Loss: {val_loss:.4f} - Val Acc: {val_acc:.4f}")
```

```python
print("PT final loss:", history_pt['train_loss'][-1])
print("PT final acc :", history_pt['train_acc'][-1])
print("PT final val loss  :", history_pt['val_loss'][-1])
print("PT final val acc   :", history_pt['val_acc'][-1])

```

```python
from sklearn.metrics import classification_report, confusion_matrix
import numpy as np

resnet.eval()
all_labels = []
all_preds = []

with torch.no_grad():
    for inputs, labels in test_loader:
        inputs = inputs.to(device)
        labels = labels.to(device)

        outputs = resnet(inputs)
        _, predicted = torch.max(outputs, 1)

        all_labels.extend(labels.cpu().numpy())
        all_preds.extend(predicted.cpu().numpy())

all_labels = np.array(all_labels)
all_preds = np.array(all_preds)

print("Classification report (PyTorch model):")
print(classification_report(all_labels, all_preds, target_names=['daisy','dandelion','rose','sunflower','tulip']))
```
